```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd 

import numpy as np

from sklearn.linear_model import LinearRegression

from sklearn.preprocessing import PolynomialFeatures
```

```python
#Multiple Linear Reegression 


#load data

df=pd.read_csv("C:\\Users\\user\\Desktop\\Machine Learning\\DataSets\\rank_salary.csv")
df=df.drop("Position",axis=1)


X=df["Level"].values.reshape(-1,1)
Y=df["Salary"].values.reshape(-1,1)

#Plotting

plt.scatter(X,Y,alpha=0.5,color="red")
plt.plot(X,Y,linewidth=2,color="blue")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()





```

```python
#as we can see from plot, this look like a polynomial,so we will use polynomial regression


poly=PolynomialFeatures(degree=2)

poly_X=poly.fit_transform(X)

Regressor2=LinearRegression()

Regressor2.fit(poly_X,Y)

y_pred=Regressor2.predict(poly_X)


```

```python
#plotting the results

plt.scatter(X,Y,alpha=0.5,color="red")
plt.plot(X,y_pred,linewidth=2,color="blue")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
# from the graph,it looks like we have a good fit,so we can now use this model
```
