```python
import seaborn as sns

import matplotlib.pyplot as plt

import pandas as pd 

import numpy as np

from sklearn.model_selection import train_test_split

from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error

from sklearn.preprocessing import StandardScaler
```

```python
#Multiple Linear Reegression 


#load data

df=pd.read_csv("C:\\Users\\user\\Desktop\\Machine Learning\\DataSets\\weatherHistory.csv")

df=df.drop("Loud Cover",axis=1)

plt.figure(figsize=(8,12))

heatmap=sns.heatmap(df.corr(),vmin=-1,vmax=1,annot=True,cmap="BrBG")

heatmap.set_title("Correlation Matrix",fontdict={"fontsize":18},pad=18)
```

```python
# we will choose Apparent Temperature and Humidity as the Features since they have the highest correlation with the dependant variable 

df=df.loc[:5000,["Temperature (C)","Apparent Temperature (C)","Humidity"]]

X=df.loc[:,["Apparent Temperature (C)","Humidity"]].values
Y=df.loc[:,"Temperature (C)"].values


#Feature Scaling

scaler=StandardScaler()
X=scaler.fit_transform(X)

#Train Test Split

X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=100)


#Model Seelction

MLR=LinearRegression()
MLR.fit(X_train,Y_train)


```

```python
y_MLRPred=MLR.predict(X_test)
```

```python
print(MLR.intercept_,MLR.coef_)
```

```python
#Accuracy check 

mean_squared_error(Y_test,y_MLRPred)
```

```python
# the Mean Squared Error is low compared to when the Regressor only had one feautre variable,which means our model fits better with two highly correlated variables instead of just 1,so we should choose this model instead.
```
