```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import seaborn as sns

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import LabelEncoder


```

# 1 - Loading The Dataset

```python
# link to dataset https://www.kaggle.com/code/prashant111/eda-logistic-regression-pca/data?select=adult.csv


df=pd.read_csv("C:\\Users\\user\\Downloads\\adult.csv")

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

<p> There are no null values in the dataset,but some features contain ? as a value, so we need to fix that</p>

```python
df[df=='?']=np.nan

df
```

```python
df.info()
```

<p> now since we have null values in the workclass,occupation, and native.country columns we can replace it with the most frequent value in that column, which is also called as the mode</p>

```python
for col in ['workclass','occupation','native.country']:
    
    df[col].fillna(df[col].mode()[0], inplace=True)
    
    
```

```python
df.info()
```

```python
df.isnull().sum()
```

<p> Now we have no null values,so we can continue</p>



# 2 - Feature Engineering

<p> We need to convert the categorical feautures to numerical feautures so our model can interpert it better,we will do this using One Hot Encoding </p>

```python
df.info()
```

```python

X=df.drop('income',axis=1)

Y=df.income

X.head()
```

```python
Y.head()
```

```python
X.head()
```

```python
categories = ['workclass', 'education', 'marital.status', 'occupation', 'relationship', 'race', 'sex', 'native.country']
for feature in categories:
        encoder= LabelEncoder()
        X[feature] = encoder.fit_transform(X[feature])
        


X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=1)




```

```python
X_train.head()
```

```python
X_train.shape
```

```python
X_test.shape
```

# 3 - Model Selection

```python
from sklearn.ensemble import RandomForestClassifier

rfClf=RandomForestClassifier(max_depth=3,random_state=1)

rfClf.fit(X_train,Y_train)

```

```python
y_pred=rfClf.predict(X_test)

y_pred
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix

accuracy_score(Y_test,y_pred)
```

```python
confusion_matrix(Y_test,y_pred)
```
